##Windows Function

```python
from pyspark.sql.window import Window
import pyspark.sql.functions as F
```

```python
wdf=spark.read.options(header="True",inferschema="True").parquet("MT cars.parquet")
wdf.show()
wdf.printSchema()
```

```python
wdf.count(),wdf.filter(F.col("hp")<60).count()
```

```python
windowsdf=Window.partitionBy("cyl").orderBy("mpg")
```

```python
wdf.withColumn("rownum",F.row_number().over(windowsdf)).show()
```

```python
wdf.withColumn("rownum",F.rank().over(windowsdf)).show()
```

```python
wdf.withColumn("preHP",F.lag("hp",1).over(windowsdf)).withColumn("postHP",F.lead("hp",1).over(windowsdf)).show()
```

```python
wdf.show()
```

```python
wdf.groupBy("cyl").min("hp").show()
wdf.groupBy("cyl").max("hp").show()
wdf.groupBy("cyl").sum("hp").show()
wdf.groupBy("cyl").avg("hp").show()
```

```python
# have to add a row to show just one line
fdf=Window.partitionBy("cyl")
wdfn=wdf.withColumn("avg",F.avg("hp").over(fdf)).withColumn("min",F.min("hp").over(fdf)).withColumn("max",F.max("hp").over(fdf)).withColumn("sum",F.sum("hp").over(fdf))
wdfn.select('cyl','sum','avg','min','max').dropDuplicates().show()
```

```python
wdfn.show()
```

```python
import pyspark.sql.functions as F
read_df_dup.filter(F.col('Country Name')=='India').show()
```

```python
#Filter with Country India and year 2000
read_df_dup.filter((F.col('Country Name')=='India')&(F.col('Year')==2000)).show()
```

```python
#Filter with Country India or year 2001
df2=read_df_dup.filter((F.col('Country Name')=='India')|(F.col('Year')==2001)).select('Country Code','Value')
```

```python
df2.orderBy('Value').show()
```

```python
read_df_dup.orderBy('Value').show()
```

### Do Filter,Groupby and aggregation in one function

```python
assi_rdd=read_df.groupBy('Country Name').agg(F.sum('Value'))
assi_rdd.filter(F.col('sum(Value)')>80).show()
```
